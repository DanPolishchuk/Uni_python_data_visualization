```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("shopping_trends.csv")
df.rename(columns=lambda x: x.lower().replace(' ', '_'), inplace=True)
```

```python
# 1. Побудуйте гістограму віку клієнтів, використовуючи Seaborn. 

plt.figure(figsize=(15, 6))

sns.histplot(df["age"], bins=20, color="darkblue")

plt.title("Client age", fontweight="bold")
plt.show()
```

```python
# 2. Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які товари покупають клієнти найчастіше. 

plt.figure(figsize=(15, 6))

sns.countplot(data=df, x="item_purchased", hue="item_purchased", palette="viridis")

plt.xticks(rotation=45)
plt.xlabel("Items")
plt.title("Items distribution among clients", fontweight="bold")
plt.show()
```

```python
# 3. Використовуйте scatter plot, щоб візуалізувати залежність між віком та сумою покупок. 

plt.figure(figsize=(15, 6))

sns.scatterplot(data=df, x="age", y="purchase_amount_(usd)", color="c", marker="o")

plt.title("Dependency between client age and purchase amount", fontweight="bold")
plt.xlabel("Client Age")
plt.ylabel("Purchase Amount $")
plt.show()
```

```python
# 4. Побудуйте графік розподілу оцінок товарів для оцінки якості.

plt.figure(figsize=(15, 6))

sns.histplot(df["review_rating"] , color='yellow', bins=25)

plt.xlabel("Rating")
plt.title("Items Rating", fontweight="bold")
plt.show()
```

```python
# 5. Створіть категоріальну діаграму, щоб відобразити розподіл покупок за статтею. 

plt.figure(figsize=(15, 6))


sns.countplot(data=df, x="gender", hue="gender", palette="viridis")

plt.title("Purchase distribution by gender", fontweight="bold")
plt.show()
```

```python
# 6. Вивчіть, чи існує зв'язок між рейтингом товару і сумою покупок за допомогою scatter plot.

plt.figure(figsize=(15, 6))

sns.scatterplot(data=df, x="review_rating", y="purchase_amount_(usd)", color="b", marker="o")

plt.title("Dependency between rating and purchase amount", fontweight="bold")
plt.xlabel("Review rating")
plt.ylabel("Purchase Amount $")
plt.show()
```

```python
# 7. Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти роблять найбільше покупок. 

plt.figure(figsize=(15, 6))

sns.countplot(data=df, x="season", hue="season", palette="viridis")

plt.title("Purchase count by season", fontweight="bold")
plt.show()
```

```python
# 8.Створіть графік розподілу розмірів одягу, щоб визначити популярність різних розмірів. 

plt.figure(figsize=(15, 6))

sns.histplot(df["size"] , color="red")

plt.title("Size popularity", fontweight="bold")
plt.show()
```

```python
# 9. Порівняйте, скільки покупок робиться за різними методами оплати за допомогою стовпчатої діаграми. 

plt.figure(figsize=(15, 6))

sns.countplot(data=df, x="payment_method", hue="payment_method", palette="viridis")

plt.title("Payment methods", fontweight="bold")
plt.show()
```

```python
# 10. Створіть scatter plot для визначення, чи існує зв'язок між кількістю попередніх покупок та сумою поточної покупки. 

plt.figure(figsize=(15, 6))

sns.scatterplot(data=df, x="previous_purchases", y="purchase_amount_(usd)", color="c", marker="o")

plt.title("Dependency between previous purchase count and current purchase amount", fontweight="bold")
plt.xlabel("Previous Purchase Count")
plt.ylabel("Purchase Amount $")
plt.show()
```

```python
# Побудуйте графік поверхні (3D surface plot) для візуалізації залежності між трьома числовими змінними з використанням бібліотеки Seaborn. 
#Налаштуйте відображення графіка, включаючи назву графіку, підписи осей і інші необхідні параметри для кращого розуміння візуалізації. 
#Додайте кольорову легенду. 


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

sns.set_style("darkgrid")
fig = plt.figure(figsize=(12, 8))

ax = fig.add_subplot(111, projection='3d')

# Create a 3D surface plot
surface = ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot: z = sin(√(x^2 + y^2))')

ax.view_init(elev=20, azim=30)

cbar = fig.colorbar(surface, shrink=0.5, aspect=10)
cbar.set_label('Z Values')

plt.show()
```
