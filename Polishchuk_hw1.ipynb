```python
#Видалити дані для всієї країни з таблиці в останньому рядку таблиці 
row_to_drop = df[df["регіон"] == "Україна"].index[0]
df = df.drop(row_to_drop)
```

```python
#Замінити пропущені дані у стовпцях середніми значеннями за стовпчиком  (метод fillna) 
for i in df.columns[1:]:
    df[i].fillna(df[i].mean(), inplace=True)
    df[i] = round(df[i], 1)
df
```

```python
#Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні 

condition = df[df["2014"] > df["2014"].mean()]
condition["регіон"]
            
```

```python
#В якому регіоні була найвища народжуваність у 2010 та 2012 роках (окремо). 

# 2010 немає в таблиці, беру 2008
highscore1 = df.loc[df['2008'].idxmax()]['регіон']
highscore2 = df.loc[df['2012'].idxmax()]['регіон']

print(f"Найвища народжуванність у 2008 році - {highscore1}\n"
      f"Найвища народжуванність у 2012 році - {highscore2}")
```

```python
#Побудуйте стовпчасту народжуваності за регіонами у 2014 році 
plt.figure(figsize=(10, 6))
plt.barh(df["регіон"], df["2014"], color='skyblue')
plt.title('Народжуваність за регіонами у 2014 році')
plt.xlabel('Регіон')
plt.ylabel('Кількість народжень')

plt.show()
```
